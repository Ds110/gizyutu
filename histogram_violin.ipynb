```python
#!/usr/bin/env python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
from numpy.random import *
import numpy as np 

data = np.random.normal(0,1,1000)

plt.figure()
plt.hist(data,bins=20,range=(-5,5))
plt.xlabel('x')
plt.ylabel('N')
plt.show()
```

```python


#!/usr/bin/env python
# -*- coding: utf-8 -*-
import matplotlib.pyplot as plt
from numpy.random import *
import numpy as np 

data = np.loadtxt('testdata.csv', delimiter=",",skiprows=True)

xLabel = ['Japanese', 'Mathematics','English','Average']

fig = plt.figure()
ax = fig.add_subplot(111)

bp = ax.violinplot(data,showmeans=True, showmedians=True,showextrema=True)
bp['cmedians'].set_color('g')

ax.set_xticks(np.arange(1, len(xLabel) + 1))
ax.set_xticklabels(xLabel,fontsize=10)
ax.set_xlim(0.25, len(xLabel) + 0.75)
plt.ylim([0,100])
plt.xlabel('Subject')
plt.ylabel('Score')
plt.title('Test Result')

plt.show()
```
